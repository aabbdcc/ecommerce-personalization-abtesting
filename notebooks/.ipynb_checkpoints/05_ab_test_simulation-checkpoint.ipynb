# A/B Test Simulation for E-commerce Recommendation Strategies

## Objective

This notebook simulates an offline A/B test to compare two recommendation strategies on historical e-commerce transaction data:

- **Variant A (Control):** Global popularity-based recommendation
- **Variant B (Treatment):** Basket co-occurrence-based recommendation

Because the dataset does not contain real recommendation exposure logs, clicks, or randomized online experiment assignments, this analysis uses historical basket behavior to build **proxy evaluation metrics**.

## Experiment Question

Does a basket co-occurrence recommendation strategy outperform a simple popularity-based strategy in terms of potential product relevance and cross-sell opportunity?


## Key Idea

For each product purchased in an order, we generate a recommendation list from each strategy and check whether the recommendation list contains other products that were also purchased in that order.

This creates an offline proxy for recommendation usefulness.


## Important Note

This is **not a real online A/B test**. It is an **offline simulation** designed to compare baseline recommendation strategies before live experimentation.

# Preparetion

## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Load data

```python
df = pd.read_csv("../data/processed/online_retail_clean.csv")
rfm = pd.read_csv("../data/processed/customer_rfm.csv")
global_popularity_revenue = pd.read_csv("../outputs/tables/global_popularity_revenue.csv")
cooccurrence = pd.read_csv(
    "../outputs/tables/product_cooccurrence.csv",
    dtype={"stockcode_x": str, "stockcode_y": str},
    low_memory=False
)

df["invoice_date"] = pd.to_datetime(df["invoice_date"])

df.head()
```

## Merge customer segements

```python
df = df.merge(
    rfm[["customer_id", "segment"]],
    on="customer_id",
    how="left"
)

df.head()
```

# Constructing an experimental evaluation sample

## Build basket table

```python
basket_df = df[["invoice_no", "customer_id", "segment", "stockcode", "description"]].drop_duplicates()
basket_df.head()
```

## Compute basket size

```python
basket_size = (
    basket_df.groupby("invoice_no")["stockcode"]
    .nunique()
    .reset_index(name="unique_products")
)

basket_size.describe()
```

## Keep only muli-product baskets

```python
eligible_orders = basket_size[basket_size["unique_products"] >= 2]["invoice_no"].tolist()

eval_df = basket_df[basket_df["invoice_no"].isin(eligible_orders)].copy()

eval_df.shape
```

## Inspect evaluation sample

```python
eval_df.head()
```

# Define two recommendation systems

## Build global popularity recommendation list

```python
global_ranked_products = global_popularity_revenue["stockcode"].astype(str).tolist()

global_ranked_products[:10]
```

## Standardize stockcode type

```python
eval_df["stockcode"] = eval_df["stockcode"].astype(str)
cooccurrence["stockcode_x"] = cooccurrence["stockcode_x"].astype(str)
cooccurrence["stockcode_y"] = cooccurrence["stockcode_y"].astype(str)
global_popularity_revenue["stockcode"] = global_popularity_revenue["stockcode"].astype(str)
```

## Create product lookup

```python
product_lookup = (
    eval_df[["stockcode", "description"]]
    .drop_duplicates()
    .dropna()
)

product_lookup.head()
```

## Define popularity recommender

```python
def recommend_popularity(target_product, global_ranked_list, top_n=5):
    recs = [p for p in global_ranked_list if p != target_product]
    return recs[:top_n]
```

## Prepare co-occurrence recommendation map

```python
left_map = cooccurrence[["stockcode_x", "stockcode_y", "cooccurrence_count"]].rename(
    columns={
        "stockcode_x": "source_product",
        "stockcode_y": "recommended_product"
    }
)

right_map = cooccurrence[["stockcode_y", "stockcode_x", "cooccurrence_count"]].rename(
    columns={
        "stockcode_y": "source_product",
        "stockcode_x": "recommended_product"
    }
)

cooccurrence_long = pd.concat([left_map, right_map], ignore_index=True)
cooccurrence_long.head()
```

## Aggregate recommendation candidates

```python
cooccurrence_ranked = (
    cooccurrence_long.groupby(["source_product", "recommended_product"], as_index=False)
    ["cooccurrence_count"]
    .sum()
    .sort_values(["source_product", "cooccurrence_count"], ascending=[True, False])
)

cooccurrence_ranked.head(20)
```

## Build dictionary for fast lookup

```python
cooccurrence_dict = {}

for source_product, sub_df in cooccurrence_ranked.groupby("source_product"):
    cooccurrence_dict[source_product] = sub_df["recommended_product"].tolist()
```

## Define co-occurrence recommender

```python
def recommend_cooccurrence(target_product, cooccurrence_mapping, top_n=5):
    recs = cooccurrence_mapping.get(target_product, [])
    recs = [p for p in recs if p != target_product]
    return recs[:top_n]
```

# Building Offline Evaluation Logic

For each item in every order:

* Treat it as the “anchor product.”
* Treat the other items in the same order as the “ground truth basket continuation.”
* Check whether the recommendation list includes any of these other items.

## Build ground-truth basket targets

```python
order_products = (
    eval_df.groupby("invoice_no")["stockcode"]
    .apply(list)
    .reset_index(name="products_in_order")
)

order_products.head()
```

## Expand into evaluation rows

```python
evaluation_rows = []

for _, row in order_products.iterrows():
    invoice_no = row["invoice_no"]
    products = row["products_in_order"]

    for anchor_product in products:
        true_other_products = [p for p in products if p != anchor_product]

        if len(true_other_products) == 0:
            continue

        evaluation_rows.append({
            "invoice_no": invoice_no,
            "anchor_product": anchor_product,
            "true_other_products": true_other_products,
            "ground_truth_size": len(true_other_products)
        })

evaluation_df = pd.DataFrame(evaluation_rows)
evaluation_df.head()
```

## Join customer/segment metadata

```python
order_meta = (
    eval_df.groupby("invoice_no", as_index=False)
    .agg(
        customer_id=("customer_id", "first"),
        segment=("segment", "first")
    )
)

evaluation_df = evaluation_df.merge(order_meta, on="invoice_no", how="left")
evaluation_df.head()
```

# Start rating

Let’s start by defining three proxy metrics that are well-suited for a baseline:

1. Hit@K  
Whether the recommendation list contains at least one actual product
2. Recall@K  
Number of actual products in the recommendation / Total number of other actual products
3. Precision@K  
Number of actual products in the recommendation / K

## Define metric function

```python
def evaluate_recommendations(recommended_items, true_items):
    recommended_set = set(recommended_items)
    true_set = set(true_items)

    hits = recommended_set.intersection(true_set)
    hit = 1 if len(hits) > 0 else 0
    precision = len(hits) / len(recommended_items) if len(recommended_items) > 0 else 0
    recall = len(hits) / len(true_set) if len(true_set) > 0 else 0

    return hit, precision, recall, len(hits)
```

## Run simulation for both variants

```python
TOP_N = 5

results = []

for _, row in evaluation_df.iterrows():
    anchor_product = row["anchor_product"]
    true_items = row["true_other_products"]

    recs_a = recommend_popularity(anchor_product, global_ranked_products, top_n=TOP_N)
    recs_b = recommend_cooccurrence(anchor_product, cooccurrence_dict, top_n=TOP_N)

    hit_a, precision_a, recall_a, hits_a = evaluate_recommendations(recs_a, true_items)
    hit_b, precision_b, recall_b, hits_b = evaluate_recommendations(recs_b, true_items)

    results.append({
        "invoice_no": row["invoice_no"],
        "customer_id": row["customer_id"],
        "segment": row["segment"],
        "anchor_product": anchor_product,
        "ground_truth_size": row["ground_truth_size"],

        "variant_a_hit": hit_a,
        "variant_a_precision": precision_a,
        "variant_a_recall": recall_a,
        "variant_a_hits_count": hits_a,

        "variant_b_hit": hit_b,
        "variant_b_precision": precision_b,
        "variant_b_recall": recall_b,
        "variant_b_hits_count": hits_b
    })

results_df = pd.DataFrame(results)
results_df.head()
```

## Inspect simulation dataset

```python
results_df.shape
```

## Overall metric comparison

```python
overall_metrics = pd.DataFrame({
    "metric": ["Hit@5", "Precision@5", "Recall@5", "Avg Hits Count"],
    "Variant A - Popularity": [
        results_df["variant_a_hit"].mean(),
        results_df["variant_a_precision"].mean(),
        results_df["variant_a_recall"].mean(),
        results_df["variant_a_hits_count"].mean()
    ],
    "Variant B - Co-occurrence": [
        results_df["variant_b_hit"].mean(),
        results_df["variant_b_precision"].mean(),
        results_df["variant_b_recall"].mean(),
        results_df["variant_b_hits_count"].mean()
    ]
})

overall_metrics
```

## Visualize overall comparison

```python
plot_df = overall_metrics.set_index("metric")

plot_df.plot(kind="bar", figsize=(10, 5))
plt.title("Offline A/B Simulation: Recommendation Strategy Comparison")
plt.ylabel("Metric Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Statistical Tests

## Paired t-test for recall

```python
t_stat_recall, p_value_recall = stats.ttest_rel(
    results_df["variant_b_recall"],
    results_df["variant_a_recall"]
)

t_stat_recall, p_value_recall
```

## Paired t-test for precision

```python
t_stat_precision, p_value_precision = stats.ttest_rel(
    results_df["variant_b_precision"],
    results_df["variant_a_precision"]
)

t_stat_precision, p_value_precision
```

## McNemar-style simplified hit comparison proxy

```python
results_df["hit_diff"] = results_df["variant_b_hit"] - results_df["variant_a_hit"]
results_df["hit_diff"].value_counts()
```

## Mean lift calculation

```python
lift_summary = pd.DataFrame({
    "metric": ["Hit@5", "Precision@5", "Recall@5"],
    "absolute_lift": [
        results_df["variant_b_hit"].mean() - results_df["variant_a_hit"].mean(),
        results_df["variant_b_precision"].mean() - results_df["variant_a_precision"].mean(),
        results_df["variant_b_recall"].mean() - results_df["variant_a_recall"].mean()
    ]
})

lift_summary["relative_lift_pct"] = [
    lift_summary.loc[i, "absolute_lift"] / overall_metrics.loc[i, "Variant A - Popularity"]
    if overall_metrics.loc[i, "Variant A - Popularity"] != 0 else np.nan
    for i in range(len(lift_summary))
]

lift_summary
```

# Analyze results by segment

```python
# Segment-level comparison
```

```python
segment_metrics = (
    results_df.groupby("segment", as_index=False)
    .agg(
        variant_a_hit=("variant_a_hit", "mean"),
        variant_b_hit=("variant_b_hit", "mean"),
        variant_a_precision=("variant_a_precision", "mean"),
        variant_b_precision=("variant_b_precision", "mean"),
        variant_a_recall=("variant_a_recall", "mean"),
        variant_b_recall=("variant_b_recall", "mean")
    )
)

segment_metrics
```

## Segment recall lift

```python
segment_metrics["recall_lift"] = segment_metrics["variant_b_recall"] - segment_metrics["variant_a_recall"]
segment_metrics["hit_lift"] = segment_metrics["variant_b_hit"] - segment_metrics["variant_a_hit"]

segment_metrics
```

## Plot recall lift by segment

```python
plt.figure(figsize=(10, 5))
plt.bar(segment_metrics["segment"], segment_metrics["recall_lift"])
plt.xticks(rotation=45, ha="right")
plt.title("Recall Lift (Variant B - Variant A) by Segment")
plt.ylabel("Recall Lift")
plt.tight_layout()
plt.show()
```

## Plot hit lift by segment

```python
plt.figure(figsize=(10, 5))
plt.bar(segment_metrics["segment"], segment_metrics["hit_lift"])
plt.xticks(rotation=45, ha="right")
plt.title("Hit@5 Lift (Variant B - Variant A) by Segment")
plt.ylabel("Hit Lift")
plt.tight_layout()
plt.show()
```

## Interpretation

The offline simulation compares a simple popularity-based recommendation baseline against a basket co-occurrence recommendation strategy.

### Main findings
- Variant A (global popularity) provides a strong baseline and is likely useful for cold-start or low-information contexts.
- Variant B (basket co-occurrence) is expected to capture stronger product relationships because it leverages observed purchase combinations.
- If Variant B shows higher Hit@5, Recall@5, or Precision@5, this suggests that basket-aware recommendation may better support cross-sell opportunities than a generic popularity ranking.

### Segment implications
- If treatment lift varies across customer segments, recommendation logic may need to be adapted by segment rather than deployed uniformly.
- Higher lift among high-value or repeat customers would support more targeted personalization investment for these groups.
- Lower lift in low-engagement segments may justify using cheaper fallback recommendation strategies.

### Business implication
This simulation provides evidence for which baseline strategy should be prioritized for future online experimentation. It also helps define which customer groups are the best candidates for live A/B testing.

## Limitations

This notebook does not measure real user exposure, click-through rate, or conversion uplift because the dataset does not contain actual recommendation logs or randomized treatment assignment.

The evaluation is based on historical basket completion behavior, which should be interpreted as an offline proxy rather than a true causal estimate.

A real production A/B test would require:
- randomized user assignment,
- recommendation exposure logs,
- click and purchase attribution,
- clear experiment windows,
- guardrail metrics such as revenue, conversion, and user retention.
